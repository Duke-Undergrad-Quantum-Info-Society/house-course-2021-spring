Review the properties of the Bloch sphere: https://www.st-andrews.ac.uk/physics/quvis/simulations_html5/sims/blochsphere/blochsphere.html

```python
coords_cartesian = [1, 0, 0] # [x, y, z]
plot_bloch_vector(coords_cartesian, coord_type='cartesian', title="Cartesian coordinates")
```

## Exercise 2

Calculate the eigenvectors and corresponding eigenvalues of the following operators:

1. $X=\begin{bmatrix}0 & 1\\1 & 0\end{bmatrix}$


2. $Y=\begin{bmatrix}0 & -i\\i & 0\end{bmatrix}$


3. $Z=\begin{bmatrix}1 & 0\\0 & -1\end{bmatrix}$


4. (optional) Hadamard:
$H=\frac{1}{\sqrt{2}}\begin{bmatrix}1 & 1\\1 & -1\end{bmatrix}$

Note: X, Y, Z, I (identity matrix) are famously known as the Pauli matrices. We will talk more about them in Lecture 4. 

Solutions:

1. The eigenvectors of $X$ are $\frac{1}{\sqrt{2}} \begin{bmatrix}1 \\ 1\end{bmatrix}$ and $\frac{1}{\sqrt{2}} \begin{bmatrix}1 \\ -1\end{bmatrix}$ with eigenvalues 1 and -1

2. The eigenvectors of $Y$ are $\frac{1}{\sqrt{2}} \begin{bmatrix}1 \\ i\end{bmatrix}$ and $\frac{1}{\sqrt{2}} \begin{bmatrix}1 \\ -i\end{bmatrix}$ with eigenvalues 1 and -1

2. The eigenvectors of $Z$ are $\begin{bmatrix}1 \\ 0\end{bmatrix}$ and $\begin{bmatrix}0 \\ 1\end{bmatrix}$ with eigenvalues 1 and -1

2. The eigenvectors of $H$ are $\frac{1}{\sqrt{4+2\sqrt{2}}} \begin{bmatrix}1+\sqrt{2} \\ 1\end{bmatrix}$ and $\frac{1}{\sqrt{4-2\sqrt{2}}} \begin{bmatrix}1-\sqrt{2} \\ 1\end{bmatrix}$ with eigenvalues 1 and -1
